```python
import json
import pandas as pd
from matplotlib import pyplot as plt

file = 'stats_myy.json'
data = None
with open(file, 'r') as f:
    data = json.load(f)
    
player_id, stats = data.popitem()
player_stats = stats["player_stats"]
match_stats = stats["match_data"]
```

```python
print("Recent results", player_stats["Recent Results"])
print("Total Wins", player_stats["Wins"])
```

```python

stats = []
for match in match_stats:
    try:
        match_id = match
        assists = int(match_stats[match_id]["Assists"])
        deaths = int(match_stats[match_id]["Deaths"])
        hs = int(match_stats[match_id]["Headshots"])
        hsp = int(match_stats[match_id]["Headshots %"])
        kills = int(match_stats[match_id]["Kills"])
        mvps = int(match_stats[match_id]["MVPs"])
        aces = int(match_stats[match_id]["Penta Kills"])
        match_map = match_stats[match_id]["Map"]
        rounds = int(match_stats[match_id]["Rounds"])
        result = int(match_stats[match_id]["Result"])
        team = match_stats[match_id]["Team"]
        stats.append([match_id, kills, deaths, assists, hs, hsp, mvps, aces, team, match_map, rounds, result])
    except:
        pass
```

```python
df = pd.DataFrame(stats, columns=["ID", "Kills", "Deaths", "Assists", "HS", "HS%", "MVPs", "5ks", "Team", "Map", "Rounds", "Result"])
df.head(50)
```

```python
df2 = df.explode('Team')
matches = df2.groupby('Team').size()
df2 = df2.groupby(['Team']).filter(lambda x:x['Team'].count()>10).groupby(['Team']).mean()
df2.rename(columns = {'Team':'Teammate'}, inplace = True)
df2['Matches'] = matches
df2.sort_values(by=['Kills'], ascending=False)

```

```python

means = df.head(50).groupby(["Map"]).mean()['Kills']

ax = means.plot(kind = 'bar')
ax.set_ylabel('Average Kills')
```
